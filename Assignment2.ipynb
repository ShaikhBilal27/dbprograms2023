6- program to accept ID, city and department.

```python
from pymongo import MongoClient
client=MongoClient("mongodb+srv://ShiaikhBilal:MH27ch8792@bilalcluster.5clinio.mongodb.net/?retryWrites=true&w=majority")
db=client["Office"]
coll=db["workers"]
ID=input('Enter Employee ID : ')
empid={}
empid['ID']=ID
print(empid)
city=input('Enter Employee City : ')
ecity={}
ecity['city']=city
print(ecity)
dept=input('Enter Employee Department: ')
empdept={}
empdept['dept']=dept
print(empdept)
city=input('Enter New City : ')
updatedcity={}
updatedcity['city']=city
print(updatedcity)
setcityas={'$set':updatedcity}
print(setcityas)
coll.update_one(ecity,setcityas)
print('Employee City Updated')
dept=input('Enter New Department : ')
updateddept={}
updateddept['dept']=dept
print(updateddept)
setdepartmentas={'$set':empdept}
print(setdepartmentas)
coll.update_one(empdept,setdepartmentas)
print('Employee Department Updated Succesfully')
```

**7 - Fetch and Delete

```python
from pymongo import MongoClient
client=MongoClient("mongodb+srv://ShiaikhBilal:MH27ch8792@bilalcluster.5clinio.mongodb.net/?retryWrites=true&w=majority")
db=client["Office"]
coll=db["workers"]
worker_id=int(input("Enter Worker ID: "))
document = coll.find_one(worker_id)
print(document)
qr={}
qr['_id']=worker_id
coll.delete_one(qr)
print('document deleted successfully')
if document:
    client.Office.exworkers.insert_one(document)
    print("Worker document moved to exworkers collection.")
else:
    print("Worker not found.")

```
